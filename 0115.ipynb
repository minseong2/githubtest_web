```python
ratio=[34,32,16,18]
labels=['Apple','Banana','Melon','Grapes']
wedgeprops={'width':0.7,'edgecolor':'k','linewidth':1}

plt.pie(ratio,labels=labels,autopct='%.1f%%',explode=[0,0.1,0,0.1],colors=['red','yellow','green','purple'],wedgeprops=wedgeprops)
plt.show()
```

```python
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales_2019=[100,120,140,110,130,140,160,170,180,200,190,210]
sales_2020=[90,110,130,120,140,160,170,160,150,180,200,190]


plt.plot(months,sales_2019,label='2019')
plt.plot(months,sales_2020,label='2020')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison (2019~2020)',loc='center')
plt.legend(loc='best',ncol=1,fontsize=10,frameon=True,shadow=True)
plt.show()
```

```python
categories=['Category 1','Category 2','Category 3','Category 4','Category 5']
data=[20,35,15,27,45]

plt.bar(x=categories,height=data)
plt.ylabel('Values')
plt.xlabel('Categories',labelpad=10)

plt.xticks(categories,rotation=45)
plt.yticks([0,10,20,30,40,50])
plt.title('Bar Chart',loc='center')
plt.grid(True,axis='both')

plt.show()
```

```python
labels=['Apple','Banana','Mango','Blueberry']
sizes=[15,30,45,10]
explode=(0,0.1,0,0)
wedgeprops={'width':1}

plt.pie(sizes,labels=labels,autopct='%.1f%%',startangle=90,explode=explode,shadow=True,wedgeprops=wedgeprops)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x_axis = np.linspace(0,10,200) # 0~10 200등분
sin_1 = np.sin(x_axis)
cos_1 = np.cos(x_axis)
plt.title('Sine and Cosine Functions', loc='center')
plt.plot(x_axis,sin_1, label = 'sin(x)',color = 'blue')
plt.plot(x_axis,cos_1, label = 'cos(x)',color = 'red')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True, axis = 'both')
plt.legend(loc = 'best', ncol =1, fontsize = 10, frameon = True, shadow = True)
plt.show()
```
